```python
# 1. 데이터 확인
import pandas as pd

happiness_df = pd.read_csv('./datasets/happiness_report_2023.csv')
display(happiness_df)
```

```python
#2. 4분위수 확인
display(happiness_df.describe().T)
```

```python
# 1인당 GDP에 따른 삶의 자율도 변화
```

```python
data = happiness_df[['region','gdp_per_capita','freedom_to_make_life_choices']]
display(data)
```

```python
# 주요분석대상인 관심변수(freedom_to_make_life_choices 와 gdp_per_capita)들의 이상치 확인하고 정상치 범위 찾기
```

```python
import numpy as np
from scipy.stats import iqr

gdp_Q1 = data.describe().T.loc['gdp_per_capita', '25%']
gdp_Q3 = data.describe().T.loc['gdp_per_capita', '75%']
print(gdp_Q1, gdp_Q3)

iqr_value = iqr(data.gdp_per_capita)
print(iqr_value)

lower_bound = gdp_Q1 - 1.5 * iqr_value
upper_bound = gdp_Q3 + 1.5 * iqr_value
print("=" * 60)
print(f'gdp_per_capita 정상치 범위: {lower_bound} ~ {upper_bound}')
print("=" * 60)

freedom_Q1 = data.describe().T.loc['freedom_to_make_life_choices', '25%']
freedom_Q3 = data.describe().T.loc['freedom_to_make_life_choices', '75%']
print(freedom_Q1, freedom_Q3)

iqr_value = iqr(data.freedom_to_make_life_choices)
print(iqr_value)

lower_bound = freedom_Q1 - 1.5 * iqr_value
upper_bound = freedom_Q3 + 1.5 * iqr_value
print("=" * 60)
print(f'freedom_to_make_life_choices의 정상치 범위: {lower_bound} ~ {upper_bound}')
print("=" * 60)
```

```python
# 정상치 범위를 벗어난 이상치를 삭제하거나 대체하고 싶은데 이를 판단하기 위하여 히스토그램으로 판단하기
```

```python
#지역별 gdp 평균값
region_mean_df = data.groupby('region').mean('gdp_per_capita').reset_index()
display(region_mean_df)

#지역별 gdp 최소값
region_min_df = data.groupby('region').min('gdp_per_capita').reset_index()
display(region_min_df)

#지역별 gdp 최대값
region_max_df = data.groupby('region').max('gdp_per_capita').reset_index()
display(region_max_df)

#지역별 gdp 표준편차
region_std_df = happiness_df.groupby('region')['gdp_per_capita'].std().reset_index()
display(region_std_df)
```

```python
#지역별 free 평균값
freedom_mean_df = data.groupby('region').mean('freedom_to_make_life_choices').reset_index()
display(freedom_mean_df)

#지역별 free 최소값
freedom_min_df = data.groupby('region').min('freedom_to_make_life_choices').reset_index()
display(freedom_min_df)

#지역별 free최대값
freedom_max_df = data.groupby('region').max('freedom_to_make_life_choices').reset_index()
display(freedom_max_df)

#지역별 free표준편차
freedom_std_df = happiness_df.groupby('region')['freedom_to_make_life_choices'].std().reset_index()
display(freedom_std_df)
```
