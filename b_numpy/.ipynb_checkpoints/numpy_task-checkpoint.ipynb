#### Numpy Task

```python
import numpy as np

# 1 ~ 10까지의 요소를 ndarray에 담는다.
ndarray1 = np.arange(start=1, stop=11)
print(ndarray1)
print("=" * 30)

# 3행 2열 list를 ndarray에 담고 각 요소는 순서대로 6, 5, 4, 3, 2, 1이다.
ndarray1 = np.arange(start=1, stop=7)
ndarray2 = np.sort(ndarray1)[::-1]
ndarray3 = ndarray1.reshape((3,2))
print(ndarray1)
print(ndarray2, ndarray2.shape)
print(ndarray3.shape)
print("=" * 30)

# 4행 1열 ndarray, 순서대로 10, 8, 6, 4를 담는다.
ndarray1 = np.arange(start=4, stop=11, step=2)
print(ndarray1)
ndarray2 = np.sort(ndarray1)[::-1]
print(ndarray2, ndarray2.shape)
ndarray3 = ndarray2.reshape((4,-1))
print(ndarray3, ndarray3.shape)
print("=" * 30)

# 1~10까지 요소를 ndarray에 담고 각 요소에 5씩 더해준다.
ndarray1 = np.arange(start=1, stop=11)
ndarray2 = ndarray1 +5
print(ndarray2)
print("=" * 30)

# axis 0이 10인 shape이면서 모든 원소가 0, dtype은 int32인 ndarray만들기
ndarray1 = np.zeros((10,), dtype=np.int32)
print(ndarray1, ndarray1.dtype)
print("=" * 30)

# axis 0이 3, axis1이 4인 shape이면서 모든 원소가 1인 ndarray 만들기
ndarray1 = np.ones((3,4))
print(ndarray1)
print("=" * 30)

# axis 0이 5, 각 요소가 0~4인 ndarray 만들기
ndarray1 = np.arange(start=0, stop=5)
print(ndarray1)

```

```python
# 아래의 ndarray1이 주어졌을 때 순서대로 문제를 해결한다.
ndarray1 = np.arange(start=0, stop=16)
print(ndarray1, ndarray1.ndim)
print("=" * 30)

# 2 Dimension, axis 1은 2로 변경
ndarray2 = ndarray1.reshape(2,-1)
# print(ndarray2, ndarray2.ndim)
ndarray2 = ndarray2.reshape(2,-1)
print(ndarray2, ndarray2.ndim)
print("=" * 30)

# 2 Dimension, axis 0은 8로 변d경
ndarray3 = ndarray1.reshape(8,-1)
print(ndarray3, ndarray3.ndim)
print("=" * 30)

# 3 Dimension으로 변경
ndarray3 = ndarray1.reshape(4, 2, -1)
print(ndarray3, ndarray3.ndim)
print("=" * 30)

# ndarray3을 axis 1이 1인 2차원 ndarray로 변환
ndarray3 = ndarray1.reshape((-1,1))
print(ndarray3, ndarray3.ndim)
print("=" * 30)

# ndarray3을 1 Dimension으로 변환
print("=" * 30)

# 1차원 ndarray 1~100까지 값 중 97값 추출하기
ndarray3 = np.arange(start=1, stop=101)
print(ndarray3[-4])
print("=" * 30)
```

```python
# 아래의 ndarray1이 주어졌을 때 순서대로 문제를 해결한다.
ndarray1 = np.arange(start=1, stop=21)
print(ndarray1, ndarray1.ndim)
print("=" * 30)

# 2차원 배열 axis 0은 4, axis 1은 5로 1~20 요소 값 중 15값 추출하기
ndarray1 = ndarray1.reshape(4, 5, -1)
print(ndarray1[2,4])
print("=" * 30)

# 2 Dimension, axis 0은 5로 변경
ndarray2 = ndarray1.reshape(5, -1)
print(ndarray2, ndarray2.ndim)
print("=" * 30)

# 2행의 4번째 숫자 출력
print(ndarray2[1,3])
print("=" * 30)

# 1~100중 짝수만 출력
ndarray1 = np.arange(start=1, stop=101)
ndarray2 = ndarray1[ndarray1 %2 ==0]
# print(ndarray1[ndarray1 %2 ==0])
print(ndarray2)
print("=" * 30)

# 위에서 구한 짝수들을 axis 0이 10인 2차원 배열로 변환 후 82~100까지 추출
ndarray3 = ndarray2.reshape(10, -1)
# print(ndarray3, ndarray3.ndim)
# print("=" * 30)
ndarray4 = ndarray3[[8,9]]
print(ndarray4)
print("=" * 30)

# 위에서 구한 요소 중 (86, 100) 좌표 구하기
print(ndarray4[0,2])
print(ndarray4[1,4])
```
