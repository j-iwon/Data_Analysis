### RFM 분석 실습
- Recency: 얼마나 최근에 구매했는가
- Frequency: 얼마나 자주 구매했는가
- Monetary: 얼마나 많은 금액을 지출했는가

### 고객 분석
-  voice_no: 송장 번호. 명사 같은. 각 작업에 고유하게 할당된 문자 'I'와 6자리 정수의 조
- customer_id: 고객번호. 명사 같은. 각 작업에 고유하게 할당된 문자 'C'와 6자리 정수의 조합
- gender: 고객 성별의 문자a열 변수
- age: 고객 연령의 양의 정수 변수
- category: 구매한 상품 카테고리의 열 변수
- quantity: 거래당 각 제품(항목)의 수량
- price: 단가. 숫자. 단위당 제품 가격은 터키 리라(TL)
- payment_method: 거래에 사용된 결제 방법(현금, 신용 카드 또는 직불 카드)의 문자열 변수
- invoice_date: 송장 날짜. 거래가 발생한 날
- shopping_mall: 거래가 이루어진 쇼핑몰 이름의 문자열 변수의 문자열 변수

```python
# 날짜 형식 변경 (5/8/2022 -> 2022/05/08)
cs_df['invoice_date'] = pd.to_datetime(cs_df['invoice_date'], format='%d/%m/%Y').dt.strftime('%Y/%m/%d')
cs_df
```

```python
# Recency 구하기
cs_df['date_point'] = pd.to_datetime(cs_df['invoice_date']).apply(lambda x: (pd.to_datetime('2023/03/08') - x).days)
cs_df
```

```python
# 마지막 구매 날짜는 2023/03/08
cs_df['invoice_date'].sort_values(ascending=False)
```

```python
import numpy as np

Q1 = cs_df.describe().T.loc['price', '25%']
Q3 = cs_df.describe().T.loc['price', '75%']

iqr = Q3 - Q1

lower_bound = Q1 - iqr * 1.5
uppper_bound = Q3 + iqr * 1.5

if lower_bound < 0:
    lower_bound = 0

print(f'정상치 범위: {lower_bound} ~ {uppper_bound}')
```

```python
# price 이상치 제거
cs_df = cs_df[cs_df['price'] <= uppper_bound]
cs_df
```

```python
# price 이상치 제거 후 index 초기화
cs_df = cs_df.reset_index(drop=True)
cs_df
```

```python
# 사용하지 않을 컬럼 제거(invoice_no)
cs_df = cs_df.drop(['invoice_no'], axis=1)
cs_df
```

```python
# Recency
sm_rfm_r = cs_df.groupby('shopping_mall')['date_point'].min().reset_index(name='Recency')
sm_rfm_r
```

```python
# Frequency
sm_rfm_f = cs_df['shopping_mall'].value_counts().reset_index(name='Frequency')
sm_rfm_f
```

```python
# 구매수량과 가격을 곱하여 total_price를 만듬
total_price = cs_df['quantity'] * cs_df['price']
display(total_price)

cs_df['total_price'] = total_price

cs_df
```

```python
# Monetary
sm_rfm_m = cs_df.groupby('shopping_mall')['total_price'].sum().reset_index(name='Monetary')
sm_rfm_m
```

```python
# 테이블 병합
sm_rfm = pd.merge(sm_rfm_r, sm_rfm_f, on='shopping_mall')
sm_rfm = pd.merge(sm_rfm, sm_rfm_m, on='shopping_mall')
sm_rfm
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
sm_rfm[['Recency', 'Frequency', 'Monetary']] = \
                                    scaler.fit_transform(sm_rfm[['Recency', 'Frequency', 'Monetary']])
```

```python
sm_rfm
```

```python
sm_rfm['Recency'] = 1 - sm_rfm['Recency']
sm_rfm
```

```python
# RFM 점수 total
sm_rfm['TotalScore'] = sm_rfm[['Recency', 'Frequency', 'Monetary']].sum(axis=1)
sm_rfm
```

```python
# 기존 dataFrame과 sm_rfm을 병합 (cs_sm_df = shopping_mall기준으로 만든 테이블)
cs_sm_df = pd.merge(cs_df, sm_rfm, on='shopping_mall')
cs_sm_df
```

```python
# Total 구한 후 필요 없는 RFM 제거
drop_columns = [
    'Recency', 'Frequency', 'Monetary'
]

cs_sm_df = cs_sm_df.drop(labels=drop_columns, axis=1)
cs_sm_df
```

```python
import numpy as np

l1, l2, l3, l4 = np.percentile(cs_sm_df['TotalScore'], [20, 40, 60, 90])
print(l1, l2, l3, l4)
```

```python
def get_level(x):
    if x <= l1:
        return 5
    if x <= l2:
        return 4
    if x <= l3:
        return 3
    if x <= l4:
        return 2
    return 1
```

```python
cs_sm_df['Level'] = cs_sm_df['TotalScore'].apply(get_level)
cs_sm_df
```

#### 각 쇼핑몰별 이용 고객 성비
- 모든 쇼핑몰이 공통적으로 여성 고객이 남성 고객보다 많은것을 볼 수 있다.

```python
payments = ['Credit Card', 'Debit Card', 'Cash']

# 사이즈
plt.figure(figsize=(20, 8))

sns.countplot(x = 'shopping_mall', data=cs_sm_df, hue='payment_method', palette='muted', hue_order=payments)

# 제목
plt.title('Payment Count', fontsize=15)

plt.show()
```

#### 각 쇼핑몰 별 결제방식
> 결제방식 부분에서도 모든 쇼핑몰이 유사한 형태를 띄고있다
> 전체적으로 현금결제가 가장 많고 다음으로 신용카드가 현금결제보다 조금 낮은 수치를 보이며, 체크카드의 수치가 상대적으로 꽤나 낮은것을 볼 수 있다.

- 튀르키예의 쇼핑몰들이다보니 전체적이게 현금결제를 주로 하는듯 하며, 다음으로는 신용카드를 많이 사용하는 것 같다.  
(튀르키예의 작은 가게나 쇼핑몰은 대부분 현금결제밖에 되지않아서 사람들이 현금을 많이 들고다닌다고 한다.)

### 마케팅 전략 제안
- 모든 매장이 공통적으로 여성 고객이 많은점을 고려하여 여성 고객의 선호도가 더 높은 제품들을 좀 더 구체적이게 조사하여  
더 다양한 물품구비 및 새로운 이벤트 홍보를 하여서 여성 고객들을 좀 더 끌어모으고,  
 기존 고객유치를 위해 포인트제도 및 등급별 혜택을 주는 전략이 좋을 것 같습니다

- 다른 매장들에 비해 상대적으로 수치가 낮은 매장들은 높은 수치를 보이는 매장들에 비해 위치적이게나 규모가 상대적으로 불리하여  
더 큰 매장들과 경쟁을 하기보다는, 남성들이 더 선호하는 제품을 많이 다루는 등 좀 다른 분야의 마케팅을 하는 전략이 좋을 것 같습니다.
